```python
# Calculador de perfil musical
edad = int(input("Ingresa tu edad: "))
le_gusta_música = input("¿Te gusta la música? (si/no): ") == "si"
le_gusta_bailar = input("¿Te gusta bailar? (si/no): ") == "si"

if edad >= 18 and le_gusta_música and le_gusta_bailar:
  print("Eres una estrella de la fiesta")
elif edad >= 18 and le_gusta_música and not le_gusta_bailar:
  print("Te gusta la música pero no tanto bailar")
else:
  print("Tienes tu propio estilo")
```

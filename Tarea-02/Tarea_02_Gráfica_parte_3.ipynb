```python
# Detector de puntaje gamer
numero1 = int(input("Ingresa primer puntaje: "))
numero2 = int(input("Ingresa segundo puntaje: "))

suma = numero1 + numero2

if suma < 50:
  print("Resultado: Nivel principiante")
elif suma >= 50 and suma < 80:
  print("Resultado: Nivel intermedio")
else:
  print("Resultado: Nivel pro")
```
